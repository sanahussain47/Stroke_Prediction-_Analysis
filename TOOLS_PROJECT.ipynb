Handle missing values (imputation)  column by column

Why: ML models require no NaNs. Choose method by column type and distribution.

BMI (continuous numeric)

Commonly many bmi are missing. Options:

Impute with median (robust to outliers).

Impute groupwise median (e.g., median by gender or work_type) if pattern exists.


```python
print("BMI missing %:", df['bmi'].isnull().mean()*100)

# Strategy: group-median by gender then global median fallback
df['bmi'] = df.groupby('gender')['bmi'].transform(lambda x: x.fillna(x.median()))
df['bmi'] = df['bmi'].fillna(df['bmi'].median())
```

```python
print("Missing values per column:\n", df.isnull().sum())
print("Number of duplicate rows:", df.duplicated().sum())
```
